```python
leveinshtein_case.loc[:,'Hamming Score']=''
leveinshtein_case
```

```python
## Calculating the Hamming Distance between 2 Strings
#original hamming_distance = hamming(list(string1), list(string2)) * len(string1)
for i in range(len(leveinshtein_case)):
    
    str1=leveinshtein_case['Original Name'].iloc[i]
    str2=leveinshtein_case['Test Case Name'].iloc[i]
    Normalized_HD_ = distance.hamming(list(str1), list(str2))
    leveinshtein_case.at[i,'Hamming Score']=Normalized_HD_*len(str1)
leveinshtein_case
#ValueError: The 1d arrays must have equal lengths.
```

### Jaro-Winkler Score
#### Question is that : the lower the Jaro–Winkler distance for two strings is, the more similar the strings are. which is opposite to other distance metrics
"Jaro-Winkler takes into account only matching characters and any required transpositions (swapping of characters). Also it gives more priority to prefix similarity. Levenshtein counts the number of edits to convert one string to another."

```python
!pip install jellyfish
import jellyfish
```

```python
leveinshtein_case.loc[:,'Jaro-Winkler Score']=''
leveinshtein_case
```

```python
name1=leveinshtein_case['Original Name'].iloc[1000]
name2=leveinshtein_case['Test Case Name'].iloc[1000]   
jellyfish.jaro_winkler(name1,name2)
```

```python
for i in range(len(leveinshtein_case)):
    name1=leveinshtein_case['Original Name'].iloc[i]
    name2=leveinshtein_case['Test Case Name'].iloc[i]                           
    leveinshtein_case.at[i,'Jaro-Winkler Score']=jellyfish.jaro_winkler(name1,name2)
leveinshtein_case
```

```python
df5 = leveinshtein_case.sort_values('Jaro-Winkler Score', ascending=False)
df5
```

```python
from Levenshtein import distance
```

```python
name1=leveinshtein_case['Original Name'].iloc[1000]
name2=leveinshtein_case['Test Case Name'].iloc[1000]   
distance(name1,name2)
```

```python
leveinshtein_case.loc[:,'Levenshtein Distance Score']=''
leveinshtein_case
```

```python
for i in range(len(leveinshtein_case)):
    name1=leveinshtein_case['Original Name'].iloc[i]
    name2=leveinshtein_case['Test Case Name'].iloc[i]                           
    leveinshtein_case.at[i,'Levenshtein Distance Score']=distance(name1,name2)
leveinshtein_case
```
