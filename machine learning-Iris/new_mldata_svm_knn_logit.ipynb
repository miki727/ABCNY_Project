```python
best_p = -1
best_score = 0.0
best_k = -1
for method in ["uniform", "distance"]:
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(n_neighbors = 3, weights = method, p = p)
        knn_clf.fit(X_train, y_train)
        knn_score = knn_clf.score(X_test, y_test)
        if knn_score > best_score:
            best_score = knn_score
            best_p = p
            best_method = method


print("best_p = ", best_p)
print("best_method = ", best_method)
print("best_score = ", best_score) 
```

```python
X=clean1[['Bridger Score', 'levenshtein_distance', 'jaro_winkler_similarity',
       'match_rating_comparison', 'partial_ratio', 'token_set_ratio',
       'w_ratio', 'matching_numbers', 'log_fuzz_score_numbers']]
Y=clean1['predict']
scale= StandardScaler()
 
# standardization of dependent variables
scaled_data = scale.fit_transform(X) 
X_train, X_test, y_train, y_test = train_test_split(scaled_data, Y, test_size=0.3,random_state=42) # 70% training and 30% test
```

```python
kernals = ['Polynomial', 'RBF', 'Sigmoid','Linear']#A function which returns the corresponding SVC model
def getClassifier(ktype):
    if ktype == 0:
        # Polynomial kernal
        return SVC(kernel='poly', degree=8, gamma="auto")
    elif ktype == 1:
        # Radial Basis Function kernal
        return SVC(kernel='rbf', gamma="auto")
    elif ktype == 2:
        # Sigmoid kernal
        return SVC(kernel='sigmoid', gamma="auto")
    elif ktype == 3:
        # Linear kernal"
        return SVC(kernel='linear', gamma="auto")
```
