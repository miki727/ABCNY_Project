```python
def month_dict():
    dic = {'Jan':'01','Feb':'02','Mar':'03',
           'Apr':'04','May':'05','Jun':'06',
           'Jul':'07','Aug':'08','Sep':'09',
           'Oct':'10','Nov':'11','Dec':'12'}
    return dic
```

```python
ofac_list_dob_filtered = ofac_list_dob.copy()
#extract date of birth
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split(';')[0])[4:])
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split('to')[0]) if re.search('to',x) else x)
for i in month_dict().keys():
    ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : re.sub(i,month_dict()[i],x) if re.search(i,x) else x)
ofac_list_dob_filtered['DOB']=ofac_list_dob_filtered['DOB'].apply(lambda x : (x.split('circa')[1]) if re.search('circa',x) else x)
```

```python
count=[]
for i in range(len(ofac_list_dob_filtered)):
    count.append(len(ofac_list_dob_filtered.iloc[i]['DOB'].split(' ')))
ofac_list_dob_filtered['count'] = ofac_list_dob_filtered['DOB'].apply(lambda x:len(x.split(' ')))
```

```python
ofac_list_dob_filtered
```

```python
ofac_list_dob_mdy = ofac_list_dob_filtered.copy()
ofac_list_dob_my = ofac_list_dob_filtered.copy()
#ofac_list_dob_mdy = ofac_list_dob_filtered[ofac_list_dob_filtered['count']==3]
ofac_list_dob_mdy['DOB']=ofac_list_dob_mdy['DOB'].apply(lambda x:re.sub(' ','/',x))
ofac_list_dob_mdy
```

```python
ofac_list_dob_y = ofac_list_dob_mdy[ofac_list_dob_mdy['DOB'].str.count('/')==2]
ofac_list_dob_y['DOB']=ofac_list_dob_y['DOB'].apply(lambda x:x[3:5]+'-'+x[0:2]+'-'+x[8:])
ofac_list_dob_y
```

```python
uid='UID-466'
theme = list(bhin[bhin['UID']==uid]['Theme'])[0]
category = list(bhin[bhin['UID']==uid]['Category'])[0]
sub_category = list(bhin[bhin['UID']==uid]['Subcategory'])[0]
entity_type = ['Individual','Vessel']
ofac_list_sampled = ofac_list_dob_y.sample(n = 10)
ofac_list_sampled['DOB']=ofac_list_sampled['DOB'].apply(lambda x:str(int(x[0:2])+random.randint(-5,5))+'-'+str(int(x[0:2])+random.randint(3,5))+'-'+str(int(x[6:])+random.randint(-5,5)))
for index, row in ofac_list_sampled.iterrows():
    original_name = row['name']
    name_pice = row['name'].split(',')
    final_test_name =''
    while len(name_pice)>0:
        n=random.choice(name_pice)
        final_test_name=final_test_name+' '+n
        name_pice.remove(n)
        final_test_name=final_test_name.upper()
    
    final_test_cases.loc[len(final_test_cases)] = [uid, theme, category, sub_category, entity_type[0], uid + ' - ' + str(index), row['uid'], row['name'], final_test_name+' '+row['DOB']] # append to the dataframe

final_test_cases
```
